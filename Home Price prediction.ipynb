```python
X = df.drop(columns=['price'])
Y = df['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0) 
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
column_transform = make_column_transformer((OneHotEncoder(sparse_output=False), ['beds']), remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
LR = LinearRegression()
```

```python
X_scaled = scaler.fit_transform(X)
LR.fit(X_scaled, Y)
```

```python
pipe = make_pipeline(column_transform, scaler, LR)
```

```python
pipe.fit(X_train, Y_train)
```

```python
Y_pred_LR = pipe.predict(X_test)
```

```python
r2_score(Y_test, Y_pred_LR)
```

# Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_transform, scaler, ridge)
```

```python
pipe.fit(X_train, Y_train)
```

```python
Y_pred_Ridge = pipe.predict(X_test)
r2_score(Y_test, Y_pred_Ridge)
```

# Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_transform, scaler, lasso)
```

```python
pipe.fit(X_train, Y_train)
```

```python
pipe.fit(X_train, Y_train)
```

```python
Y_pred_Lasso = pipe.predict(X_test)
r2_score(Y_test, Y_pred_Lasso)
```

```python
print("No Regularization: ", r2_score(Y_test, Y_pred_LR))
print("Lasso: ",r2_score(Y_test, Y_pred_Lasso))
print("Ridge: ",r2_score(Y_test, Y_pred_Ridge))
```
